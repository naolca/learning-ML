# Importing the necessary packages

```python

import matplotlib.pyplot as plt
import numpy as np
import random
```

# prepare the dataset

```python
features = np.array([1,2,3,5,6,7])
labels = np.array([155, 197, 244, 356,407,448])

print(features)
print(labels)
```

# Lets plot the points


```python
def plot_points(features, labels):
    X = np.array(features)
    y = np.array(labels)
    plt.scatter(X, y)
    plt.xlabel('number of rooms')
    plt.ylabel('prices')

def draw_line(slope, y_intercept, color='grey', linewidth=0.7, starting=0, ending=8):
    x = np.linspace(starting, ending, 1000)
    plt.plot(x, y_intercept + slope*x, linestyle='-', color=color, linewidth=linewidth)

plot_points(features, labels)
    
```

# 1. square trick method to update the price per room(m) and base price (y-intercept or bias)

```python
def square_trick(actual_price, price_per_room, number_of_rooms, base_price, learning_rate):

    pridicted = price_per_room*number_of_rooms + base_price
    error = actual_price - pridicted

    new_price_per_room = price_per_room + error * learning_rate * number_of_rooms
    new_base_price = base_price + error * learning_rate
    
    return new_price_per_room, new_base_price

```

# 2. Absolute trick method to update the price per room(m) and base price (y-intercept or bias)

```python
def absolute_trick(actual_price, price_per_room, number_of_rooms, base_price, learning_rate):
    
        pridicted = price_per_room*number_of_rooms + base_price
        error = actual_price - pridicted
    
        if error > 0:
            new_price_per_room = price_per_room + learning_rate * number_of_rooms
            new_base_price = base_price + learning_rate
        else:
            new_price_per_room = price_per_room - learning_rate * number_of_rooms
            new_base_price = base_price - learning_rate
        return new_price_per_room, new_base_price
```
